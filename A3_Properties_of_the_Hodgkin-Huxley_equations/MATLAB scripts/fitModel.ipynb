This code is to fit a model to the second threshold as a function of time between interstimulus currents

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.array([6,8,10,12,14,16,18,20])
y = np.array([145.293,71.6016,40.7129,25.4219,16.9219,12.7188,11.2813,11.5469])
t = np.linspace(5,21,100)
```

```python
plt.plot(x,y,'o')
plt.xlim(5,21)
plt.xlabel("Time")
plt.ylabel("Threshold to the second pulse ")
plt.show()
```

```python
best_error = 100000
best_A = 0
best_Alpha = 0
best_b = 0

for b in range(0,12):
    for k in range(0,15000):
        a=k/10
        for alpha in range(1,100):
            al = alpha / 100
            temp = b+a*np.exp(-al*x)
            error = np.sqrt(sum((temp-y)**2))
            if error<best_error:
                best_error = error
                best_A = a
                best_Alpha = al
                best_b = b

print("Done")
```

Therefore the solution model is, $threshold(t) = 1394 e^{-0.39t} + 11$

```python
fig, ax = plt.subplots(figsize=(10,10))
ax.plot(t,best_b+best_A*np.exp(-best_Alpha*t))
ax.plot(x,y,'o')
ax.set_xlabel('time')
ax.set_ylabel('current amplitude')
ax.set_title("$Model --> threshold(t) = 1394 e^{-0.39t} + 11$")
plt.show()
```
